```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def trajectory(t, Xi1, Xi2):
    return np.sin(t) * Xi1 + np.cos(t) * Xi2
```

```python
def fun1(t, A, B):
    return np.sqrt(A**2 + B**2) * np.sin(t + np.arccos(A / np.sqrt(A**2 + B**2)))
```

```python
def fun2(t, A, B):
    return np.sqrt(A**2 + B**2) * np.sin(t + np.arctan(A / B))
```

```python
def fun3(t, A, B):
    return np.sqrt(A**2 + B**2) * np.sin(t + np.arcsin((A - B) / np.sqrt(A**2 + B**2)))
```

```python
t = np.linspace(0, 50, 101)

Xi1 = 1
Xi2 = 2

plt.plot(t, trajectory(t, Xi1, Xi2), '-')
plt.plot(t, fun1(t, Xi1, Xi2), '-')
plt.plot(t, fun2(t, Xi1, Xi2), '-')
plt.plot(t, fun3(t, Xi1, Xi2), '-')
```
